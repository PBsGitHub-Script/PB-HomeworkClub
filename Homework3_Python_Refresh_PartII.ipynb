#### Q.1. Create a documented python script that would

a. take this number 112345678911234566 and count the number of 2s in the string and print out the number.

b. take a sentence from user input, turn it all to lowercase letters and remove the spaces and count the length and print out the length. You choose the sentence.

```python
#Python script to count number of 2s in a given string

I= ('112345678911234566') #enter the number as a string
print (list(I)) #list each character in a string
count=0
for num in list(I): #use for loop control flow statement 
    if num=='2':    #use if else statement to count the number only when the value is 2
        count=count+1  #count the number is the value matches
    else:               
        pass           #else pass the character in the string
print('number of 2s is', count,'.') #print out the number of 2s in the given string
```

```python
#Python script to input a sentence, change it to lower case sentence, and count the length without spaces

A = input ("Input your sentence here:") #input any sentence you like
B = A.lower()                          #changes it to lowercase sentence
C = (B.replace(' ',''))                #removes the spaces in between the words to change it to letters
print("Your lowercase letters are:",C,".") #print your lowercase letters
Count_length = len(C)                   #count the length of your letters
print ("Length of your sentence without spaces is",Count_length,".") #print the length of your letters
```
